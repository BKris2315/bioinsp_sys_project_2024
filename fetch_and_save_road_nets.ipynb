```python
import osmnx as ox
```

```python
query = [("beijing", "Beijing", "China"),
 ('tokyo', 'Tokyo', 'Japan'),
 ('taipei', 'Taipei', 'Taiwan'),
 ('viena', 'Viena', 'Austria')]

query = [("beijing", "Beijing, China"),
 ('tokyo', 'Tokyo, Japan')]
```

```python
for city, city_name in query:
    # Define the city name


    # Download the road network for driving
    G = ox.graph_from_place(city_name, network_type='drive')

    # Convert graph to GeoDataFrames
    nodes, edges = ox.graph_to_gdfs(G)

    # Save nodes and edges to CSV
    nodes.reset_index().to_csv(f"road_nets/{city}_nodes.csv", index=False)
    edges.reset_index().to_csv(f"road_nets/{city}_edges.csv", index=False)

    # Plot the road network
    fig, ax = ox.plot_graph(G, save=True, filepath=f"imgs/{city}_road_network.png", dpi=300)
```
